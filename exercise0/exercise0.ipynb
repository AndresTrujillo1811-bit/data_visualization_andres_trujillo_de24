```python
co2_emissions = duckdb.query("""SELECT year, month, AVG(Avg_CO2) AS CO2 FROM df GROUP BY year, month ORDER BY year DESC""").df().set_index("Month")
co2_emissions
```

```python
results_co2 = co2_emissions.loc[1].plot(title="Cars CO2 used in Norway 2007-2016", x="Year", y="CO2", xlabel="Year", ylabel="CO2 used", style="--o")
results_co2
```

```python
import matplotlib.pyplot as plt 

x = co2_emissions["Year"].loc[1]
y = co2_emissions["CO2"].loc[1]

plt.title("Cars average CO2 used in Norway 2007-2016")
plt.xlabel("Years from 2007")
plt.ylabel("CO2 used")
plt.plot(x, y, marker='o')
plt.show()
```
