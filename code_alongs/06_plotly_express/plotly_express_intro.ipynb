# Plotly express

```python
import plotly.express as px

gapminder = px.data.gapminder()
gapminder.head()
```

```python
gapminder.info()
```

```python
nordic = gapminder[gapminder["country"].isin(["Sweden", "Norway", "Denmark", "Finland", "Iceland"])]
nordic.head()
```

```python
nordic["year"].value_counts()
```

```python
nordic[nordic["year"] == 2007]
```

```python
sweden = nordic.query("country == 'Sweden'")
sweden
```

```python
nordic2007 = nordic.query("year == 2007")
nordic2007
```

# Bar chart

```python
fig = px.bar(nordic2007, x ="country", y = "gdpPercap", title = "GDP per capital in nordic countries 2007")
fig.show()
fig.write_html("interactive_figures/gpd_nordic_2007_bar.html")
```
